```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

SEABRON HAS TWO TYPE OF FIGURES
1. figure
2. axes

6 TYPES OF FIGURES
1. relational plots
2. categorical plots
3. distribution plots
4. matrix plots
5. regression plots
6. multi plots

<h1>RELATIONAL PLOTS

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',color='red',hue='sex',style='time',size='size') #axes function 
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',color='red',marker='+') #figure function relplot
```

LINE PLOT

```python
gap = px.data.gapminder()
gap.head()
```

```python
temp_df = gap[gap['country']=='India']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df.sample()
```

```python
#multiple lines in line plot
sns.relplot(data=temp_df,x='year',y='lifeExp',hue='country',kind='line',size='continent')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',hue='country',kind='scatter',size='continent')
```

FACET PLOT -> only works with figure level plots

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='sex',col='time')
```

```python
#col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year')
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```

Distribution plots

displot -> figure level funtion

histplot , kdeplot, rugplot -> axes level function

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',hue='sex',element='step') #to view without lines
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(data=titanic,x='age',kind='hist')
```

```python
sns.displot(data=titanic,x='age',kind='hist',col='survived',element='step',hue='sex') #male and female survivors and deceased
```

KDEPLOTS (Kernel Density Estimation (KDE)) used for representing probability

```python
sns.kdeplot(data=titanic,x='age')
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',height=10)#use height and aspect to control figure level plots
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)#fill parameter to fill color in the plot
```

```python
#rugplot is always used as a compliment to other plots
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')#rugplot shows ticks of data points on graph
```

```python
#histogram representing bivariate data -> two data
sns.histplot(data=tips,x='total_bill',y='tip') #below graph shows that 15 and 3 are the most frequent combination
```

```python
#similarly kde bivariate plot 
sns.kdeplot(data=tips,x='total_bill',y='tip') #below graph shows if the lines are closer then the density is more, distant lines shows low density
```

Matrix plots
1. heatmap (only has axes level function)
2.clustermap

```python
gap.head()
```

```python
temp_df = gap.pivot_table(index='country',columns='year',values='lifeExp')
```

```python
#since heatmap is a axes level function we can use figsize to change size
plt.figure(figsize=(20,20))
sns.heatmap(temp_df)
```

```python
#annot=True to show the annotation on the heatmap depicting the value of each grid

temp_df = gap[gap['continent'].isin(['Europe'])].pivot(index='country',columns='year',values='lifeExp')
```

```python
#annot example
plt.figure(figsize=(20,20))
sns.heatmap(temp_df,annot=True,linewidth=0.5) #linewidth=creates gap between grids
```

Clustermaps uses machine learning algo internally to plot the similarity between cols

```python
df = px.data.iris()
```

```python
plt.figure(figsize=(40,43))
sns.clustermap(df.iloc[:,[0,1,2,3]])
```
