```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
plt.style.available
```

```python
plt.style.use('Solarize_Light2')
```

```python
tips= sns.load_dataset('tips')
```

<h1>Categorical plots

<h3>Categorical scatter plot
1.Strip plot ( basically a scatter plot but in bar plot form )
2.Swarm plot
<h3>Categoical distribution plot
1.Boxplot
2.Violinplot
<h3>Categorical Estimate plot -> for central tendency
1.barplot
2.pointplot
3.countplot

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y="total_bill",kind='strip',jitter=False)#jitter to make in one line
```

```python
sns.catplot(data=tips,x='day',y="total_bill",kind='strip',jitter=1,hue='sex') #isnt useful as no proper representation of data
```

```python
#swarmplot -> gives better representation of the data

sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill')
```

Categorical distribution plots


```python
#boxplot
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
#single varibale box plot
sns.boxplot(data=tips,y='total_bill')
```

Violin plot ( combination of boxplot and kdeplot )

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

Categorical estimate plot -> used for central tendency

```python
#barplot 
sns.barplot(data=tips,x='sex',y='total_bill')#errorbar shows the range in which the values can range
```

```python
#remove errorbar
sns.barplot(data=tips,x='sex',y='total_bill',errorbar=None)
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='bar',hue='smoker') #y value is mean of total bill
```

```python
#using estimator parameter we can run any funnction
import numpy as np
sns.catplot(data=tips,x='sex',y='total_bill',kind='bar',estimator=np.std)#using numpy to run standard deviation
```

```python
#point plot -> draw line between errorbar points to show the difference
sns.pointplot(data=tips,x='sex',y='total_bill')
```

```python
#countplot
sns.countplot(data=tips,x='sex')
```

```python
sns.catplot(data=tips,x='sex',hue='day',kind='count')
```

```python
#facet in catplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='bar',col='time',row='smoker')
```

<h2>Multi grid plots

```python
#FacetGrid
g = sns.FacetGrid(data=tips,row='time',col='day',hue='smoker',height=10)
g.map(sns.barplot,'sex','total_bill')
g.add_legend()
```

```python
#pairplot
import pandas as pd
iris = pd.read_csv('datasets/iris.csv')
sns.pairplot(iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm']])
```

```python
sns.pairplot(iris,hue='Species')
```

```python
#pairgrid -> gives more customization option for pairplot
g = sns.PairGrid(data=iris,hue='Species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue='Species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(data=iris,hue='Species',vars=['SepalWidthCm','PetalWidthCm'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

Joint plot and JointGrid plot

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
g= sns.JointGrid(tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.histplot)
```

```python
g= sns.JointGrid(tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
