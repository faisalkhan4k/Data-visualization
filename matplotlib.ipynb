```python
df = pd.read_csv('datasets\sharma-kohli.csv')
```

```python
df.head()
```

```python
plt.plot(df['index'],df['V Kohli'],color='green')
plt.plot(df['index'],df['RG Sharma'],color='red')

plt.title('Rohit vs Kohli')
plt.xlabel('years')
plt.ylabel('runs')
```

```python
plt.plot(df['index'],df['V Kohli'],color='green',linestyle= 'dashed')
plt.plot(df['index'],df['RG Sharma'],color='red',linestyle='dotted')

plt.title('Rohit vs Kohli')
plt.xlabel('years')
plt.ylabel('runs')
```

```python
#line(style and width ) + marker(style and size)
plt.plot(df['index'],df['V Kohli'],color='green',linestyle= 'dashed',linewidth=4,marker='D',markersize=10)
plt.plot(df['index'],df['RG Sharma'],color='red',linestyle='dashdot', linewidth=3,marker='o',markersize=15)

plt.title('Rohit vs Kohli')
plt.xlabel('years')
plt.ylabel('runs')
```

```python
plt.plot(df['index'],df['V Kohli'],label='kohli',color='pink',linewidth=1,marker='*')
plt.xlabel('years')
plt.ylabel('runs')
plt.legend()
```

```python
plt.plot(df['index'],df['V Kohli'],label='kohli',color='pink',linewidth=1,marker='*')
plt.xlabel('years')
plt.ylabel('runs')
plt.ylim(400,1000)#y axis limit
plt.xlim(2009,2015)# x axis limit (range)
plt.legend()
```

```python
#multiple bars in a barchart
df= pd.read_csv(r'datasets\batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]),df['2015'])
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
#plt.bar(df['batsman'],df['2016'])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.1)#moving the bar to the left of x-axis by 0.2
plt.bar(np.arange(df.shape[0])-0.1,df['2016'],width=0.1)#adding the next one in the new place created
plt.bar(np.arange(df.shape[0]),df['2017'],width=0.1)
plt.xticks(np.arange(df.shape[0]),df['batsman'],rotation=90)
```

STACKED BAR CHART

```python
plt.bar(df['batsman'],df['2017'],label='2017')#first
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')#keeping 2017 add 2016
plt.bar(df['batsman'],df['2015'],bottom=(df['2017']+df['2016']),label='2015')#keeping 2016 and 2017 add 2015
plt.legend()
plt.show()
```

```python
df=pd.read_csv('datasets\gayle-175.csv')
df
```

Adding custom colors and using the explode paramter in pie charts

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],colors=['red','pink','orange','lightblue','beige','black'],explode=[0.2,0,0,0,0,0],shadow=True)
```

```python
#important
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')# autopct important
```

Saving and sharing plots using plt.savefig

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')# autopct important
plt.savefig('matplotlib-plot1.png')
```
