```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets\iris.csv')
df.sample(5)
```

```python
df['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2},inplace=True)
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap='jet')#other options for cmap are summr, winter, jet and many more
```

```python
plt.figure(figsize=(25,7))

plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap='winter',alpha=0.5) #alpha is opacity
plt.colorbar()
```
