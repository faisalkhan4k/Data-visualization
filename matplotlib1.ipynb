```python
df = pd.read_csv(r'datasets\batter.csv')
df = df.head(100).sample(25,random_state=42)
df
```

```python
plt.figure(figsize=(25,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'])
plt.xlabel('average')
plt.ylabel('strike rate')

for i in range(df.shape[0]):
    if(df['batter'].iloc[i]=='V Kohli'):
        plt.text(df['avg'].iloc[i],df['strike_rate'].iloc[i],df['batter'].iloc[i],fontdict={'size':15,'color':'brown'})
    else:

        plt.text(df['avg'].iloc[i],df['strike_rate'].iloc[i],df['batter'].iloc[i],fontdict={'size':15,'color':'green'})

plt.show()
```

Horizontal and vertical lines on graph

```python
plt.figure(figsize=(25,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'],c=df['runs'],cmap='summer')
plt.legend()

for i in range(df.shape[0]):
    plt.text(df['avg'].iloc[i],df['strike_rate'].iloc[i],df['batter'].iloc[i],fontdict={'color':'red'})

plt.axhline(140,color='blue')#horizontal line
plt.axvline(35,color='green')
plt.axvline(30,color='orange')#vertical line
plt.colorbar()
plt.grid()
plt.show()
```

SUBPLOTS

```python
fig,ax= plt.subplots(nrows=2,ncols = 1,figsize=(15,10))#sharex parameter for same x axis 


ax[0].scatter(df['avg'],df['strike_rate'],color='red',marker='D',s=5)
ax[0].set_title('Avg vs SR')
ax[0].set_ylim(140,170)#all functions work the same, just requires the set_ in the front
ax[0].set_ylabel('SR')


ax[1].hist(df['runs'],bins=[0,1000,2000,3000,10000])
ax[1].set_title('Runs')
ax[1].set_ylabel('Occurences')

```

2D subplots

```python
fig,ax = plt.subplots(nrows=2,ncols=2)

# ax[0].hist(df['runs']) shows error because its a 2D array of axes
ax[0][0].hist(df['runs'])
ax[0][1].hist(df['runs'])
ax[1][0].hist(df['runs'])
ax[1][1].hist(df['runs'])
```

2nd method of making subplots

```python
fig =plt.figure(figsize=(25,17))#create figure object and then keep adding subplot while adjusting the rows and columns

ax1 = fig.add_subplot(2,3,1) # first two numbers is rows and columns and the third number represents the figure
ax1.hist(df['runs'])

ax2 = fig.add_subplot(2,3,2)
ax2.pie(df['runs'],autopct='%0.1f%%',labels=df['batter'])

ax3= fig.add_subplot(2,3,3)
ax3.bar(df['batter'],df['avg'])
ax3.set_ylabel('average')
ax3.set_xticks(np.arange(df.shape[0]),df['batter'],rotation=90)
ax3.axhline(30,color='red')

ax4=fig.add_subplot(2,3,4)
ax4.pie(df['runs'])

fig.show()
```

3D SURFACE PLOTS

```python
x = np.linspace(-10,10,100) # 100 equidistance points from -10 to 10
y = np.linspace(-10,10,100) # 100 equidistance points from -10 to 10
```

```python
x.shape
```

```python
XX , YY = np.meshgrid(x,y)# it create 2d array from 1d array, if x has 100 rows and y has 300 values now x shape will be (300,100),here since both shape are 100 the resulting shape is 100,100
```

```python
XX.shape
```

```python
z = XX**2 + YY**2 #using a 3d function to plot a surface plot
```

```python
z.shape
```

```python
fig =plt.figure(figsize=(12,8))

ax=plt.subplot(projection='3d')

p = ax.plot_surface(XX,YY,z,cmap='jet')
fig.colorbar(p)#colorbar function works different here from other plots 
```

```python
#another example

z = np.sin(XX) + np.cos(YY)
fig =plt.figure(figsize=(12,8))

ax=plt.subplot(projection='3d')

p = ax.plot_surface(XX,YY,z,cmap='winter')
fig.colorbar(p)
```

CONTOUR PLOTS

```python
#contour plot is a 2d version of a 3d plot
fig,ax = plt.subplots(nrows=1,ncols=2,subplot_kw={'projection': '3d'},figsize=(15,12))#if we want 3d plots in subplots

ax1=fig.add_subplot(1,2,2,projection='3d')
ax1.plot_surface(XX,YY,z,cmap='winter')

ax2=fig.add_subplot(1,2,1)
ax2.contourf(XX,YY,z,cmap='winter')#if its jsut contour then its not visually appealing
```

```python
z = np.sin(XX) - np.cos(YY)

plt.contour(XX,YY,z)
```

```python
z = np.sin(XX) - np.cos(YY)

plt.contourf(XX,YY,z)
```
